```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Sort the dataframe by state and year
df.sort_values(['name', 'year'], inplace=True)

# Calculate the year-over-year growth rate for each state
df['income_growth_rate'] = df.groupby('name')['income'].pct_change()

# Pivot the dataframe to get years as columns and states as rows with income growth rate as values
df_pivot = df.pivot(index='name', columns='year', values='income_growth_rate')

# Plot a line chart for each state
plt.figure(figsize=(10,6))
for state in df_pivot.index:
    plt.plot(df_pivot.columns, df_pivot.loc[state], label=state)

plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.legend(loc='best')
plt.title('Year-Over-Year Income Growth Rate for Each State')
plt.show()
'''
This code will plot a line chart showing the year-over-year income growth rate for each state from 2005 to 2019. Each state is represented as a separate line in the chart.
'''
```
